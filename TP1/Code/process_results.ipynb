```python
import os
```

```python
'''Carrega palavras (target) para lista  words. '''

words = []

with open(f"words.txt", "r") as text_file:
        for line in text_file:
            words.append(line)
```

```python
def generate_results(analogies_dataset="analogies_cbow_25_5.txt", model="cbow_25_5.bin", prefix="results_cbow_25_5.txt", words=''):
    '''Le arquivo de saida das analogias e pega resposta, alem disso, pega os indices das analogies validas.'''

    analogies = []
    indexes = []

    with open(f"{analogies_dataset}", "r") as text_file:
        for index, line in enumerate(text_file):
            w = line.split(' ')
            if len(w)  == 6:
                indexes.append(index)
                analogies.append(w[4])
    
    valid_words = [word.split('\n')[0]  for index, word in enumerate(words) if index in indexes]
    
    correct = 0
    filtered_words = []
    filtered_analogies = []
    for index, word in enumerate(valid_words):
        if word == analogies[index]:
            correct = correct + 1
        else:
            filtered_words.append(word)
            filtered_analogies.append(analogies[index])
            
    with open(f"teste2.txt", "w") as text_file:
        for index, word in enumerate(filtered_words):
            text_file.write(f'{word} {filtered_analogies[index]}\n')
        text_file.write(f'EXIT')
    
    print(f'FILTERED WORDS {len(filtered_words)}')
    print(f'FILTERED ANALOGIES {len(filtered_analogies)}')
    
    n = len(filtered_words)
    os.system(f"./distance {model} {n} > {prefix}")
    
    n_words = len(words)
    
    out_of_dict = 0
    similarity = 0
    with open(f"{prefix}", "r") as text_file:
        for line in text_file:
            wds = line.split(' ')
            if len(wds) == 2:
                similarity = similarity + float(wds[1])
            else:
                out_of_dict = out_of_dict + 1

    similarity = similarity + correct

    mean_similarity = similarity/(len(valid_words) - out_of_dict)
    
    return(mean_similarity)
    ## Pega os resultados
    ## Conta quantos OUTS tem
    ## Conta quantos validos tem
    ## Soma todas os valores (correct + resultados)
    ## Divide a soma pelo valor total valido
    ## Retorna esse valor
    ## Como calcular quantas palavras estao fora do dicionario? Retornar esse valor tamb√©m!
    
```
